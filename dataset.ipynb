```python
import zipfile
with zipfile.ZipFile('/content/photos.zip', 'r') as zip_ref:
    zip_ref.extractall('/content')

os.mkdir('/content/formatedphotos')
```

```python
import PIL
import os.path
from PIL import Image, ImageFile

ImageFile.LOAD_TRUNCATED_IMAGES = True

f = r'/content/photos/'
for folder in os.listdir(f):
  print(folder)
  for file in os.listdir(f+'/'+folder+'/'):
    print(file)
    f_img = f+"/"+folder+'/'+file
    img = Image.open(f_img)
    img = img.convert("RGB")
    img = img.resize((300,300))
    if file.endswith(".jpeg") or file.endswith(".webp"): 
      file_name, file_ext = os.path.splitext(file)
    img.save('/content/formatedphotos/'+folder+'/{}.png'.format(file_name))
```
